#**Training and Test Data**

```python
from sklearn.model_selection import train_test_split

X1=midterm1_x
y1=midterm1_y

(X_train1, X_test1, y_train1, y_test1) = train_test_split(X1, y1, test_size=0.2, random_state=0) #random state used for reproducibility

X2=midterm2_x
y2=midterm2_y

(X_train2, X_test2, y_train2, y_test2) = train_test_split(X2, y2, test_size=0.2, random_state=0)

Xf=final_x
yf=final_y

(X_trainf, X_testf, y_trainf, y_testf) = train_test_split(Xf, yf, test_size=0.2, random_state=0)
```

#**Linear Regression**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train1, y_train1)
```

```python
from matplotlib import pyplot
import plotly.graph_objects as go
from sklearn.datasets import make_regression
importance = lr.coef_

for i,v in enumerate(importance):
 print('Feature: %0d, Score: %.5f' % (i,v))

fig = go.Figure(go.Bar(
    x=[f'Feature {i}' for i in range(len(importance))],
    y=importance,
    marker=dict(color='rgba(50, 171, 96, 0.6)'),
    name='Feature Importance'
))

fig.update_layout(
    title='Feature Importance',
    xaxis=dict(title='Feature'),
    yaxis=dict(title='Importance')
)

fig.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_lr_train_pred = lr.predict(X_train1)
y_lr_test_pred = lr.predict(X_test1)

print("Train:", r2_score(y_train1, y_lr_train_pred))
print("Test:", r2_score(y_test1, y_lr_test_pred))

# Training Set
lr_train_mse = mean_squared_error(y_train1, y_lr_train_pred)
lr_train_r2 = r2_score(y_train1, y_lr_train_pred)

# Testing Set
lr_test_mse = mean_squared_error(y_test1, y_lr_test_pred)
lr_test_r2 = r2_score(y_test1, y_lr_test_pred)
```

```python
import plotly.graph_objects as go
import numpy as np

coefficients = np.abs(lr.coef_)

feature_names = X_train1.columns
sorted_indices = np.argsort(coefficients)[::-1]
sorted_coefficients = coefficients[sorted_indices]
sorted_feature_names = feature_names[sorted_indices]

trace = go.Bar(
    x=sorted_feature_names,
    y=sorted_coefficients,
    marker=dict(color='skyblue')  )

layout = go.Layout(
    title='Linear Regression Coefficients',
    xaxis=dict(title='Feature'),
    yaxis=dict(title='Coefficient Magnitude'))

fig = go.Figure(data=[trace], layout=layout)

fig.show()
```
