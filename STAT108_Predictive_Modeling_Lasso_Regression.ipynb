#**Training and Test Data**

```python
from sklearn.model_selection import train_test_split

X1=midterm1_x
y1=midterm1_y

(X_train1, X_test1, y_train1, y_test1) = train_test_split(X1, y1, test_size=0.2, random_state=0) #random state used for reproducibility

X2=midterm2_x
y2=midterm2_y

(X_train2, X_test2, y_train2, y_test2) = train_test_split(X2, y2, test_size=0.2, random_state=0)

Xf=final_x
yf=final_y

(X_trainf, X_testf, y_trainf, y_testf) = train_test_split(Xf, yf, test_size=0.2, random_state=0)
```

#**Lasso Regression Cross Validation Midterm 1**

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LassoCV
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, r2_score
seed = 13
kfold = KFold(n_splits=5, shuffle=True, random_state=seed)

lasso1_cv = LassoCV(cv=kfold)

lasso1_cv.fit(X_train1, y_train1)
lasso1_predictions_train = lasso1_cv.predict(X_train1)
print("Train:", r2_score(y_train1, lasso1_predictions_train))
lasso1_predictions_test = lasso1_cv.predict(X_test1)
print("Test:", r2_score(y_test1, lasso1_predictions_test))
print("Alpha:", lasso1_cv.alpha_)

lasso1_cv_train_mse = mean_squared_error(y_train1, lasso1_predictions_train)
lasso1_cv_train_r2 = r2_score(y_train1, lasso1_predictions_train)
lasso1_cv_test_mse = mean_squared_error(y_test1, lasso1_predictions_test)
lasso1_cv_test_r2 = r2_score(y_test1, lasso1_predictions_test)
```

```python
lasso1_cv_performance_data = [['Midterm 1 Lasso Regression CV', lasso1_cv_train_mse, lasso1_cv_train_r2, lasso1_cv_test_mse, lasso1_cv_test_r2]]
midterm_all_performance = pd.DataFrame(lasso1_cv_performance_data, columns=['Model', 'Training MSE', 'Training R2 Score', 'Testing MSE', 'Testing R2 Score'])
midterm_all_performance
```
