#**Load the Dataset**

The data from the 4 files were seperated into 4 additional csv files. These files are seperated by homework, lecture and activity, projects, and midterms. The csv files are then concatenated into one dataframe.

```python
import pandas as pd
hw_df = pd.read_csv("https://raw.githubusercontent.com/trishaha/STAT108-Student-Success-Dashboard/main/Fall_23_24_HW.csv")
lecture_df = pd.read_csv("https://raw.githubusercontent.com/trishaha/STAT108-Student-Success-Dashboard/main/Activity_Lec_F23_F24.csv")
midterm_df = pd.read_csv("https://raw.githubusercontent.com/trishaha/STAT108-Student-Success-Dashboard/main/Midterms_F23_F24.csv")
projects_df = pd.read_csv("https://raw.githubusercontent.com/trishaha/STAT108-Student-Success-Dashboard/main/Projects_Clean.csv")
df = pd.concat([hw_df, lecture_df, midterm_df, projects_df], axis=1)
df
```

#**Midterm 1 Features**
HW 1: Vocab

HW 2: Qualitative Data

HW 3: Quantitative Displays

HW 4: Quantitative Data part 1

HW 5: Quantitative Data part 2

HW 6: Relative Position

HW 7: Midterm 1 Review

Project 1



```python
#midterm 1 features
midterm1_x = df.loc[:,[
    "HW 1: Vocab ",
    "HW 2: Qualitative Data ",
    "HW 3: Quantitative Displays ",
    "HW 4: Quantitative Data part 1",
    "HW 5: Quantitative Data part 2",
    "HW 6: Relative Position",
    "Project 1 (combine)",
    "Project 2 (combine)",
    "Project 3 (combine)",
    "Activity Week 1",
    "Activity Week 2",
    "Activity Week 3",
    "Activity Week 4",
    "Activity Week 5",
    "Lecture 1",
    "Lecture 2",
    "Lecture 3",
    "Lecture 4",
    "Lecture 5",
    "Lecture 6",
    "Lecture 7",
    "Lecture 8",
    "Lecture 9",
    "Lecture 10",
    "Lecture 11"
]]

midterm1_y = df['Midterm 1']
```

```python
from sklearn.model_selection import train_test_split

X1=midterm1_x
y1=midterm1_y

(X_train1, X_test1, y_train1, y_test1) = train_test_split(X1, y1, test_size=0.2, random_state=0) #random state used for reproducibility
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LassoCV
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, r2_score
seed = 13
kfold = KFold(n_splits=5, shuffle=True, random_state=seed)

lasso1_cv = LassoCV(cv=kfold)

lasso1_cv.fit(X_train1, y_train1)
lasso1_predictions_train = lasso1_cv.predict(X_train1)
print("Train:", r2_score(y_train1, lasso1_predictions_train))
lasso1_predictions_test = lasso1_cv.predict(X_test1)
print("Test:", r2_score(y_test1, lasso1_predictions_test))
print("Alpha:", lasso1_cv.alpha_)

lasso1_cv_train_mse = mean_squared_error(y_train1, lasso1_predictions_train)
lasso1_cv_train_r2 = r2_score(y_train1, lasso1_predictions_train)
lasso1_cv_test_mse = mean_squared_error(y_test1, lasso1_predictions_test)
lasso1_cv_test_r2 = r2_score(y_test1, lasso1_predictions_test)
```
